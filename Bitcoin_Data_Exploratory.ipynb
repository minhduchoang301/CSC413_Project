```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('/content/bitcoin_data.csv')
df
```

```python
# Check missing values
print(df.isnull().sum())

# Remove rows with null values
df.dropna(inplace=True)
```

```python
df['converted_ts'] = df['Timestamp'].apply(lambda x: datetime.datetime.fromtimestamp(x))
```

```python
df
```

```python
df.set_index("converted_ts", inplace = True)
# Resampling to daily and monthly frequency
df_day = df.resample("D").mean()
```

```python
df_day
```

```python
df_day.to_csv("bitcoin_daily.csv")
```

```python
df_month = df.resample("M").mean()
```
